## Viso Geral dos Tipos Nativos em Python

Os tipos nativos, so as categorias de objetos que esto disponveis para o desenvolvedor a partir da instalao bsica do Python, sem qualquer adio de novos pacotes.

Os tipos bsicos podem ser simples ou compostos.

Tipos simples so objetos unitrios, com apenas um valor armazenado em um endereo nico na memria.

Tipos compostos so colees formadas por outros objetos, sejam objetos unitrios, ou mesmo outros objetos compostos. Trata-se de objetos que ocupam diversos endereos na memria e que so organizados de forma a que possamos acessar os diversos valores neles contidos.

* **Simples**
    * **Integers** (nmeros inteiros, ou int)
        * 1    20    123    9    55555 <br><br>
    * **Floats** (nmeros decimais, ou float point number)
        * 1.0   20.0  123.0000000   9.0  55555.000   3.333333333333333   66.66666666666    1e-2   1e-3 <br><br>
    * **Booleans**  (boleanos, ou bool, valores lgicos Verdadeiro e Falso)
        * True False
        * Em operaes matemticas, True recebe o valor 1, enquanto False recebe o valor zero <br><br>
    * **None** (Valor Nulo. No confundir com 0, ou vazio. Trata-se de um valor com suas propriedades especficas)
    
<br>

* **Compostos**
    * **Strings** (str, ou cadeia de caracteres, so colees imutveis de caracteres delimitadas por aspas duplas ou simples)
        * "abcd", "Isto  uma string", "Cada uma das letras desta string  salva em um lugar distinto da memria"
        * True e False dentro de uma string no tm valor lgico, pois so uma mera coleo de caracteres
        * 1 + 1 dentro de uma string no gera resultado, pois so meros caracteres sem valor matemtico
        * os valores de uma string so acessveis por ndice posicional (indexao), com primeiro elemento contado a partir do nmero zero. <br><br> 
    * **Lists** (Listas so colees mutveis de objetos separados por vrgulas e delimitados por chaves \[ \] \, podendo receber, ao mesmo tempo, valores de diversos tipos, inclusive compostos)  
        * [1, 2, 'trs', 'quatro'], [['uma string contida dentro de uma lista que est contida em outra'], [1,2,3,4], [None]]  <br><br>   
    * **Tuples** (tuplas so colees imutveis delimitadas por parnteses que contm objetos separados por vrgulas.
        * (100, 1000, 10000)  <br><br>
    * **Dictionaires** (dicionrios, ou dict, so coleo de pares chave-valor (key:value). Os pares so separados entre si por vrgulas e as colees so delimitadas por colchetes {}. As chaves devem ser imutveis, enquanto os valores podem ser quaisquer tipos de objetos, mesmo compostos, inclusive outros dicionrios.
        * {a:1, b:2, c:3}
        * {"primeira lista":[1,2,3,4], "segunda lista":[5,6,7,8], "primeiro dicionrio": {a:1, b:2, c:3}} <br><br>
    * **Sets** (conjuntos so colees de objetos separados por vrgulas e delimitados por colchetes, que no admitem duplicidade de valores internos)
        * {1, 2, 3, 4, ~~1~~}
        * embora tambm sejam delimitados por colchetes, no h como confundir os sets com os dicionrios, pois dicionrios, em seu interior, apresentam pares key:value.
        * os valores inseridos duplicadamente em um set so automaticamente ignorados
        * os valores em um set no so acessveis por ndice posicional.  preciso iterar pelo conjunto para acessar seus valores, um-a-um.

### Slicing

Podemos aproveitar o tema das listas para avanar sobre outra questo que no analisamos anteriormente, embora fosse aplicvel tambm s strings, o _slicing_, ou fatiamento. _Slicing_  a recuperao de valores em intervalos indexados. No intervalo, indicamos a partir de que posio queremos obter dados, inclusive, e at que posio, exclusive. Separamos esses dois argumentos por dois pontos ":".
vejamos o exemplo com uma lista que contenha as letras do alfabeto.

````python
list_alf = list(alfab)
print(list_alf)
````

````python
lista_de_listas = [list_alf, [1,2,3,4,5,6,7,8,9]]
````

````python
print(lista_de_listas)
````

````python
lista_de_listas[0][-2:]
````

No exemplo acima, a posio 0  a letra "a", enquanto a posio 5  a letra "f". O fatiamento recuperou cinco letras, desde "a" at "e", que est na posio 4 da lista. Como podemos ver, o fatiamento inclui o primeiro valor indicado, mas exclui o ltimo.

Um mtodo muito importante ligado s tuplas  chamado de _tuple unpacking_. Trata-se de atribuir os valores constantes em uma tupla a variveis, em um nico processo. Vejamos um exemplo adiante, com uma tupla que guarda as coordenadas geogrficas do Centro de Curitiba:

````python
coordenadas_curitiba = (-25.4284, -49.2733)
````

````python
lat, long = coordenadas_curitiba
````

````python
print(lat)
````

````python
print(long)
````

Note que os valores constantes dentro da tupla passaram a integrar valores das variveis lat e long com apenas uma operao de atribuio.

Suponha que, precisando apenas da latitude e da longitude, a tupla de origem dos dados ainda desse outros dois valores, altitude e populao. Como poderamos fazer a atribuio das variveis lat e long, deixando de lado a altitude?

Isso pode ser feito, utilizando _underscore_ "_" como uma varivel descartvel, que recebe os valores de altitude e de populao.

````python
coordenadas_curitiba = (-25.4284, -49.2733, 935, 1933105)
````

````python
lat, long, _, _ = coordenadas_curitiba
````

````python
print(lat)
````

````python
print(long)
````

Precisamos utilizar o artifcio acima, pois o nmero de variveis  esquerda do operador de atribuio deve ser igual o nmero itens contidos na tupla.

### Sets

Sets, ou conjuntos, so tipos de dados compostos por valores no ordenados que no podem ser repetidos. Essas colees apresentam grande utilidade para operaes como identificao de pertencimento, interseo, unio etc. De maneira geral, sets no permitem indexao e fatiamento.



````python
set_quadrados = set((1, 2, 2, 4, 4, 9, 9, 9, 25, 25, 25, 25, 25, 25, 25, 25)) #declarao de set via construtor de objeto
````

````python
engineers = set(['John', 'Jane', 'Jack', 'Janice'])
programmers = set(['Jack', 'Sam', 'Susan', 'Janice'])
managers = set(['Jane', 'Jack', 'Susan', 'Zack'])
````

````python
employees = engineers | programmers | managers           # unio
employees
````

````python
engineering_management = engineers & managers            # interseo
engineering_management
````

````python
fulltime_management = managers - engineers - programmers # difference
fulltime_management
````

Lista	| sequncia que pode conter qualquer objeto, inclusive compostos |	Mutvel |	Indexao |	list() |	[ ]
Dicionrio | par chave-valor |	Chave imutvel, valor mutvel |	Chaveamento |	dict() |	{ }
tupla |	 uma lista imutvel |	imutvel |	Indexao |	tuple() | ( )
Set | conjunto no ordenado e sem duplicidade de elementos |	mutvel	| iterao | set() | { }



\* tipagem dinmica  mais do que isso, mas nos atenhamos somente a esse aspecto por ora.
